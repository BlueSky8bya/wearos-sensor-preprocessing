```python
import os
import pandas as pd
from datetime import timedelta

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 0. í™˜ê²½ ì„¤ì •
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
BASE_DIR = "/content/drive/MyDrive/ê°€ì²œëŒ€/RA/[wearable] ê³¼ì œ/HRV ë¶„ì„/Datas/MongoDB ë°ì´í„°"

# ì²˜ë¦¬í•  UID ë¦¬ìŠ¤íŠ¸ë¥¼ ì—¬ê¸°ì— ì‘ì„±í•˜ì„¸ìš”
UIDs = [
    # "WCbACPaBNuhUi313NtAp4bMr4xX2",
    # "SF3OKBppJOYdGucR2vffiNF0sJO2",
    # "jt6QwpW77iQgiKRAOTbFUJGofQg2",
    # "wqD6BnkBH7SHuKc70SdVYG4i4nA3",
    "loQKPAEgx1bBAvGop4cr9HqfCwn2",
    "vFjJEGZ10TZjDlJwCn5lDXQFDx42",
    "KPqggsTAIIVIYjdqiakRwzSI3I42",
    "X5MBIpEoL5clm96GCz9vKgq0byq1",
    "EykFY0NKgJVZRfI8fGCMoZ4NPc22",
    "FG6WOuZMWtgArSWGCi9MIgu4KVo1",
    "96AOPONmdGeAwZNwMtXcRumiCqV2",
    "OR2z2fOOaIM8M2obfI4YvSd7Cz03",
    "H5xpe8R0pMQNybyFmD6fpptb5s12",
    "aEbEJ4v5ZXOGXpRj4con8TY5Oks1",
    "OtvWlOKX52Zi1Qyky0yYrFxvEGT2",
    # "tZknnycQKNdJYvmp3OSSNLY6Ku92",
    # "JHXGbmkWxYOrH4mKT31aXvOaU1I3",
    # "baXYZZcSdeVQOJiogHgziaoNAbn1",
]

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 1. ì£¼ì°¨ ë¼ë²¨ë§ í•¨ìˆ˜ (cutoffs: [(ì£¼ì°¨, tz-aware Timestamp), â€¦])
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def assign_week_label(ts, cutoffs):
    for wk, cutoff_ts in cutoffs:
        if ts < cutoff_ts:
            return wk - 1   # ì˜ˆ: cutoff ì£¼ì°¨ 1ì´ë©´ ts<cutoff â†’ 0ì£¼ì°¨
    return cutoffs[-1][0]  # ë§ˆì§€ë§‰ ê²½ê³„ë¥¼ ë„˜ìœ¼ë©´ ë§ˆì§€ë§‰ ì£¼ì°¨

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 2. PPG ë°ì´í„° ì „ì²˜ë¦¬ í•¨ìˆ˜ (ìˆ˜ì •ë¨)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def preprocess_ppg_data(df):
    """
    PPG ë°ì´í„°ë¥¼ ì „ì²˜ë¦¬í•˜ì—¬ í•„ìš”í•œ ì»¬ëŸ¼ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
    - time/timestampê³¼ data/value ì»¬ëŸ¼ì„ í‘œì¤€í™”
    - ë‘ ê°€ì§€ ì»¬ëŸ¼ëª… íŒ¨í„´ì„ ëª¨ë‘ ì§€ì›: (time, data) ë˜ëŠ” (timestamp, value)
    """
    df_clean = df.copy()

    # 1. ì‹œê°„ ì»¬ëŸ¼ í™•ì¸ ë° í‘œì¤€í™” (time ë˜ëŠ” timestamp)
    time_col = None
    if 'time' in df_clean.columns:
        time_col = 'time'
    elif 'timestamp' in df_clean.columns:
        time_col = 'timestamp'
    else:
        raise ValueError("ì‹œê°„ ì»¬ëŸ¼('time' ë˜ëŠ” 'timestamp')ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

    # 2. ë°ì´í„° ì»¬ëŸ¼ í™•ì¸ ë° í‘œì¤€í™” (data ë˜ëŠ” value)
    data_col = None
    if 'data' in df_clean.columns:
        data_col = 'data'
    elif 'value' in df_clean.columns:
        data_col = 'value'
    else:
        raise ValueError("ë°ì´í„° ì»¬ëŸ¼('data' ë˜ëŠ” 'value')ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

    print(f"    ì‚¬ìš©í•  ì»¬ëŸ¼: ì‹œê°„={time_col}, ë°ì´í„°={data_col}")

    # 3. ì»¬ëŸ¼ëª… í‘œì¤€í™” (time, dataë¡œ í†µì¼)
    if time_col != 'time':
        df_clean['time'] = df_clean[time_col]
    if data_col != 'data':
        df_clean['data'] = df_clean[data_col]

    # 4. timeê³¼ dataê°€ ëª¨ë‘ ìˆëŠ” í–‰ë§Œ ìœ ì§€
    before_dropna = len(df_clean)
    df_clean = df_clean.dropna(subset=['time', 'data'])
    after_dropna = len(df_clean)
    print(f"    ê²°ì¸¡ê°’ ì œê±°: {before_dropna} â†’ {after_dropna} ({before_dropna - after_dropna} í–‰ ì œê±°)")

    # 5. ì¤‘ë³µëœ ì‹œê°„ ì œê±°
    before_dedup = len(df_clean)
    df_clean = df_clean.drop_duplicates(subset=['time'])
    after_dedup = len(df_clean)
    print(f"    ì¤‘ë³µ ì‹œê°„ ì œê±°: {before_dedup} â†’ {after_dedup} ({before_dedup - after_dedup} í–‰ ì œê±°)")

    return df_clean

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 3. í•œ UID ì²˜ë¦¬ í•¨ìˆ˜
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def process_one_uid(uid):
    print(f"[{uid}] ì²˜ë¦¬ ì‹œì‘...")

    # (a) í´ë” ì¤€ë¹„
    out_dir = os.path.join(BASE_DIR, uid, "processed")
    os.makedirs(out_dir, exist_ok=True)

    # (b) response.csv â†’ last_update_map ìƒì„± (ë¯¸êµ­ì‹œê°„ â†’ í•œêµ­ì‹œê°„ ë³€í™˜)
    resp_path = os.path.join(BASE_DIR, uid, "response.csv")
    try:
        df_resp = pd.read_csv(resp_path, encoding="cp949")
    except:
        try:
            df_resp = pd.read_csv(resp_path, encoding="utf-8")
        except Exception as e:
            print(f"[{uid}] response.csv ì½ê¸° ì‹¤íŒ¨: {e}")
            return None

    # lastUpdateë¥¼ ë¯¸êµ­ì‹œê°„ìœ¼ë¡œ íŒŒì‹± í›„ í•œêµ­ì‹œê°„ìœ¼ë¡œ ë³€í™˜
    df_resp["last_ts"] = (
        pd.to_datetime(df_resp["lastUpdate"], utc=True)
          .dt.tz_convert("Asia/Seoul")
    )

    last_update_map = {
        wk: df_resp.loc[df_resp["week"] == wk, "last_ts"].iloc[0]
        for wk in sorted(df_resp["week"].unique())
    }
    print(f"[{uid}] ì£¼ì°¨ë³„ ê²½ê³„ ì‹œê°„: {last_update_map}")

    # (c) ì£¼ì°¨ ê²½ê³„(cutoffs) ìƒì„±: 1ì£¼ì°¨ ì´ìƒë¶€í„°
    cutoffs = [
        (wk, last_update_map[wk])
        for wk in sorted(last_update_map.keys())
        if wk > 0
    ]

    # (d) watch_ppg_green.csv ì „ì²˜ë¦¬
    ppg_path = os.path.join(BASE_DIR, uid, "watch_ppg_green.csv")
    try:
        df_ppg = pd.read_csv(ppg_path, low_memory=False)
        print(f"[{uid}] ì›ë³¸ PPG ë°ì´í„° shape: {df_ppg.shape}")
        print(f"[{uid}] ì›ë³¸ ì»¬ëŸ¼: {list(df_ppg.columns)}")
    except Exception as e:
        print(f"[{uid}] PPG íŒŒì¼ ì½ê¸° ì‹¤íŒ¨: {e}")
        return None

    # 1) ê¸°ë³¸ ì „ì²˜ë¦¬ (time/timestamp, data/value ê¸°ì¤€)
    try:
        df_ppg = preprocess_ppg_data(df_ppg)
        print(f"[{uid}] ê¸°ë³¸ ì „ì²˜ë¦¬ ì™„ë£Œ")
    except Exception as e:
        print(f"[{uid}] ê¸°ë³¸ ì „ì²˜ë¦¬ ì‹¤íŒ¨: {e}")
        return None

    # 2) UTC â†’ Asia/Seoul tz-aware ë³€í™˜ (ë¯¸êµ­ì‹œê°„ â†’ í•œêµ­ì‹œê°„)
    try:
        df_ppg["time_kr"] = (
            pd.to_datetime(df_ppg["time"], utc=True)
              .dt.tz_convert("Asia/Seoul")
        )
        print(f"[{uid}] ì‹œê°„ëŒ€ ë³€í™˜ ì™„ë£Œ (UTC â†’ Asia/Seoul)")
    except Exception as e:
        print(f"[{uid}] ì‹œê°„ëŒ€ ë³€í™˜ ì‹¤íŒ¨: {e}")
        return None

    # 3) week ë¼ë²¨ë§ (0ë¶€í„° ì‹œì‘)
    if cutoffs:
        df_ppg["week"] = df_ppg["time_kr"].apply(lambda ts: assign_week_label(ts, cutoffs))
        print(f"[{uid}] ì£¼ì°¨ ë¼ë²¨ë§ ì™„ë£Œ")
        print(f"[{uid}] ì£¼ì°¨ë³„ ë°ì´í„° ë¶„í¬: {df_ppg['week'].value_counts().sort_index()}")
    else:
        df_ppg["week"] = 0
        print(f"[{uid}] ê²½ê³„ê°€ ì—†ì–´ ëª¨ë“  ë°ì´í„°ë¥¼ 0ì£¼ì°¨ë¡œ ì„¤ì •")

    # 4) uid ì»¬ëŸ¼ ì¶”ê°€
    df_ppg["uid"] = uid

    # 5) ìµœì¢… ì»¬ëŸ¼ ì„ íƒ ë° ì •ë¦¬: uid, type, time_kr, data, week
    final_columns = ['uid']

    # type ì»¬ëŸ¼ ì²˜ë¦¬
    if 'type' in df_ppg.columns:
        final_columns.append('type')
    else:
        df_ppg['type'] = 'PpgGreen'  # ê¸°ë³¸ê°’ ì„¤ì •
        final_columns.append('type')

    # ë‚˜ë¨¸ì§€ í•„ìˆ˜ ì»¬ëŸ¼ë“¤
    final_columns.extend(['time_kr', 'data', 'week'])

    # ìµœì¢… ë°ì´í„°í”„ë ˆì„ ìƒì„±
    df_ppg_final = df_ppg[final_columns].copy()

    # ì‹œê°„ìˆœ ì •ë ¬
    df_ppg_final = df_ppg_final.sort_values('time_kr').reset_index(drop=True)

    # (e) íŒŒì¼ ì €ì¥
    save_path = os.path.join(out_dir, "watch_ppg_green_week_labeled.csv")
    df_ppg_final.to_csv(save_path, index=False, encoding="utf-8-sig")
    print(f"[{uid}] ì²˜ë¦¬ ì™„ë£Œ â†’ {save_path}")
    print(f"[{uid}] ìµœì¢… ë°ì´í„° shape: {df_ppg_final.shape}")
    print(f"[{uid}] ìµœì¢… ì»¬ëŸ¼: {list(df_ppg_final.columns)}")

    # (f) ìµœì¢… ë°ì´í„° ë¶„í¬ í™•ì¸
    print(f"[{uid}] ìµœì¢… ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°:")
    print(df_ppg_final.head())
    print(f"[{uid}] ë°ì´í„° íƒ€ì… ì •ë³´:")
    print(df_ppg_final.dtypes)
    print(f"[{uid}] ì£¼ì°¨ë³„ ë°ì´í„° ê°œìˆ˜:")
    print(df_ppg_final['week'].value_counts().sort_index())
    print("-" * 80)

    return df_ppg_final

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 4. ì „ì²´ UIDs í•œêº¼ë²ˆì— ì‹¤í–‰
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def main():
    print("PPG ë°ì´í„° ì „ì²˜ë¦¬ ì‹œì‘")
    print(f"ì²˜ë¦¬í•  UID ìˆ˜: {len(UIDs)}")
    print("=" * 80)

    success_count = 0
    failed_uids = []

    for i, uid in enumerate(UIDs, 1):
        print(f"ì§„í–‰ë¥ : {i}/{len(UIDs)}")
        try:
            result = process_one_uid(uid)
            if result is not None:
                success_count += 1
            else:
                failed_uids.append(uid)
        except Exception as e:
            print(f"[{uid}] ì˜ˆì™¸ ë°œìƒ: {e}")
            failed_uids.append(uid)

    print("=" * 80)
    print("ì „ì²˜ë¦¬ ì™„ë£Œ!")
    print(f"ì„±ê³µ: {success_count}/{len(UIDs)}")
    if failed_uids:
        print(f"ì‹¤íŒ¨í•œ UIDë“¤: {failed_uids}")

if __name__ == "__main__":
    main()
```
